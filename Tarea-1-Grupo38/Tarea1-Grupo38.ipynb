## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Usuarios"

```python
%%sql
CREATE TABLE usuarios(
    id_usuario INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    apellido VARCHAR(45),
    correo VARCHAR(45) NOT NULL,
    contraseña VARCHAR(50) NOT NULL,
    fecha_registro TIMESTAMP NOT NULL,
    PRIMARY KEY (id_usuario)
);
```

### Tabla "Creadores"

```python
%%sql
CREATE TABLE creadores(
    id_creador INTEGER GENERATED ALWAYS AS IDENTITY ,
    nombre VARCHAR(45) NOT NULL,
    apellido VARCHAR(45) NOT NULL,
    popularidad INTEGER,
    estilo VARCHAR(45),
    PRIMARY KEY (id_creador)
);
```

### Tabla "Tipos_NTF"

```python
%%sql
CREATE TABLE tipos_nft(
    id_tipo INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(20) NOT NULL,
    PRIMARY KEY (id_tipo)
); 
```

### Tabla "NFT"

```python
%%sql
CREATE TABLE nft (
    id_nft INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    id_creador INTEGER NOT NULL,
    id_tipo INTEGER,
    PRIMARY KEY (id_nft),
    FOREIGN KEY (id_creador) REFERENCES creadores (id_creador),
    FOREIGN KEY (id_tipo) REFERENCES tipos_nft (id_tipo)
);
```

### Tabla "Precio NFT"

```python
%%sql
CREATE TABLE precio_nft(
    id_nft INTEGER,
    fecha TIMESTAMP NOT NULL,
    precio INTEGER NOT NULL,
    PRIMARY KEY (id_nft, fecha),
    FOREIGN KEY (id_nft) REFERENCES nft (id_nft)
);
```

### Tabla "Usuario tiene NFT"

```python
%%sql
CREATE TABLE usuario_tiene_nft (
   id_usuario INTEGER,
   id_nft INTEGER,
   PRIMARY KEY (id_usuario, id_nft),
   FOREIGN KEY (id_usuario) REFERENCES usuarios (id_usuario),
   FOREIGN KEY (id_nft) REFERENCES nft (id_nft)
);
```

### Consulta 2
**Cantidad de precios que ha tenido históricamente cada NTF.**

Ejemplo:


| NFT     | Cantidad de precios |
|--------------|----------|
| Minecraft Sphere |    3     |
| The Binary Girl     |    7     |
| ...          |   ...    |

**Solución:** Para la consulta se necesitó la unión de la tabla nft con la de precio_nft utilizando un INNER JOIN. Para poder ir contando, se utilizo la función de agregación COUNT en nft.nombre, así cada vez que se repitiera el nombre del nft relacionado a un precio no constante, se le sumara al contador, es decir, se conto cada vez que se repitiera un nombre para poder saber la cantidad de vez que el nft tuvo un precio distinto.

```python
%%sql
SELECT
	(nft.nombre) as "NFT", COUNT(nft.nombre) as "Cantidad de Precios"
FROM 
	nft INNER JOIN precio_nft
ON 
	nft.id_nft = precio_nft.id_nft
GROUP BY 
	nft.nombre;
```

### Consulta 3
**El nombre y tipo (nombre del tipo) del NFT históricamente más caro.**
###### Si hay más de uno, ponga solamente el primero
Ejemplo:

| NFT    | Tipo | 
|----------------|--------|
| Minecraft Sphere| Arte 3D |

**Solución:** Se seleccionan las variables que funcionan como los nombres de cada columna, estos eson en cuestion el nombre del NFT y el Tipo, esto se logro ya que al unir las tablas "tiposnft" y "nft" que solamente coincidan en sus casillas, tambien se verifico aquellos NFT que son Null, despues de ello se ordenaron las filas dependiendo del valor del precio de mayor a menor, al final con el "fetch" se selecciono solamente el Nft con el precio mas alto.

```python
%%sql
SELECT
    "NFT" , "Tipo"
FROM (
	SELECT
		nft.id_nft, nft.nombre as "NFT", tipos_nft.nombre as "Tipo"
	FROM
		 tipos_nft left join nft
	ON
		tipos_nft.id_tipo = nft.id_tipo
    )as tablita LEFT JOIN precio_nft
ON
	tablita.id_nft = precio_nft.id_nft
WHERE
	precio_nft.precio > 0
ORDER BY precio_nft.precio DESC
FETCH FIRST 1 ROWS ONLY;
```

**Nombre y apellido del usuario con más número de NFTs.**

```python
%%sql
SELECT 
    nombre, apellido
FROM
    (SELECT
        usuarios.nombre, usuarios.apellido, usuario_tiene_nft.id_usuario, COUNT(usuario_tiene_nft.id_usuario) as contador
    FROM 
    usuario_tiene_nft INNER JOIN usuarios 
    ON 
    usuario_tiene_nft.id_usuario = usuarios.id_usuario
    GROUP BY 
    usuarios.nombre,usuarios.apellido, usuario_tiene_nft.id_usuario
    ORDER BY
    contador DESC) as usuario_mayor_nft
FETCH FIRST 1 ROWS ONLY;
```

### Consulta 9 
**Creador al cual le han comprado más NFTs.**

Ejemplo:
   
|Nombre |Apellido |Cantidad |
|-|--|-----|
| Levi |Ackerman |32|

**Solución:** Se selecciona los todos los valores de las columnas id_usuario, id_nft de la tabla usuario_tiene_nft y todos los valores de la columna id_creador de la tabla nft, todo a partir de un INNER JOIN que fusiona todas las columnas relacionadas entre ambas tablas (usuario_tienen_nft y nft). A partir de esto, se crea una tabla, la cual es llamada "tablita". Se realiza un INNER JOIN entre tablita y la tabla creadores, y se seleccionan todos los valores de las columnas nombre y apellido de la tabla creadores. Tambien se toma nft.id_creador, que anteriormente se nombró "creador", de "tablita" para usar la función COUNT, para contar las veces que se repite y se nombre a la columna como "Cantidad". La tabla resultante se ordena de forma descendiente a partir de los valores de "Cantidad" y se limita la tabla a solo 1 fila, mostrando la consulta esperada.

```python
%%sql
SELECT
	creadores.nombre as "Nombre", creadores.apellido as "Apellido", COUNT(tablita.creador) as "Cantidad"
FROM
	(SELECT 
	usuario_tiene_nft.id_usuario, usuario_tiene_nft.id_nft, nft.id_creador as creador
	FROM
	usuario_tiene_nft INNER JOIN nft
	ON
	nft.id_nft = usuario_tiene_nft.id_nft) as tablita INNER JOIN creadores
ON
	tablita.creador = creadores.id_creador
GROUP BY
	"Nombre","Apellido"
ORDER BY 
	"Cantidad" DESC
FETCH FIRST 1 ROWS ONLY;
```

### Consulta 10  
**El NFT con mayor precio actual, que pertenezca a un creador X.**

Ejemplo:
   
|NFT |Valor|
|-|-------|
|Minecraft Sphere |18000|

**Solución:** Para solucionar esta consulta, se debió crear un WITH con una tabla temporal donde se comenzó consultando la intersección de creadores y nft para tomar el ID de cada uno y el nombre del nft con la condición de que solo sean de X creador (Tabla: creador_nft), luego se consultó con la tabla anteriormente resultante intersectada con precio_nft para tomar, además de las columnas de la primera tabla, los precios y los años. Finalmente en el SELECT principal se crea una tabla del NFT con su Valor (precio) usando la tabla temporal (tabla creador_nft_precio) cuando el año sea la fecha maxima de creador_nft_precio, luego se ordena descendientemente en base al precio y se extrae el primer valor para saber cual es el nft con mayor precio actual del creador X.

```python
%%sql
WITH creador_nft_precio as (SELECT
								"ID1","NOMBRE","ID2","NFT",precio_nft.precio as "Valor", EXTRACT(YEAR FROM precio_nft.fecha) as "Fecha"
							FROM
								(SELECT 
									creadores.id_creador as "ID1", creadores.nombre as "NOMBRE", nft.id_nft as "ID2", nft.nombre as "NFT"
								FROM
									creadores INNER JOIN nft
								ON 
									creadores.id_creador = nft.id_creador
								WHERE
									creadores.nombre = 'INSERTE X CREADOR') as creador_nft INNER JOIN precio_nft
							ON
								"ID2" = precio_nft.id_nft)
SELECT 
    "NFT","Valor"
FROM
	creador_nft_precio
WHERE
	"Fecha" = (SELECT MAX("Fecha") FROM creador_nft_precio)
ORDER BY
	"Valor" DESC
FETCH FIRST 1 ROWS ONLY
```

### Consulta BONUS
###### (Vale por un abrazo psicológico)

**El usuario que posee más dinero en NFTs considerando los precios actuales de estos**

Ejemplo:
   
|Usuario |Dinero|
|-|-------|
|Levi Ackerman |8000001|

**Solución:** Para poder resolver este problema, se creo un WITH comenzando desde la consulta anidada más interna hasta la mas externa. Primero se realizo la combinación de tablas (filtro_usuarios) usuarios y usuario_tiene_nft (INNER JOIN) para poder tener el id del usuario que tiene nft y su nombre, luego se combinó con la tabla nft (usuarios_con_nft) con INNER JOIN para poder saber cual usuario tiene cual nft creando una tabla que contenga el id del usuario y nft mas los nombres respectivos, después se combinó la tabla "usuarios_con_nft" con la de precio_nft para extraer los años y asi finalmente poder crear la tabla final creada en el SELECT principal, donde se usa la tabla temporal "fechas" tanto en el FROM como en el WHERE (para sacar la fecha actual) y asi poner en la tabla la suma de los precios de los NFT actuales comprados por el Usuario con más dinero.

```python
%%sql
WITH fechas as (SELECT 
							"Usuario", EXTRACT(YEAR FROM precio_nft.fecha) as "Fecha", precio_nft.precio as precio
						FROM 
							(SELECT 
								"ID", "Usuario", "ID NFT", nft.nombre "NFT"
							FROM
								(SELECT 
									 usuario_tiene_nft.id_usuario as "ID", CONCAT(usuarios.nombre, ' ', usuarios.apellido) as "Usuario", usuario_tiene_nft.id_nft as "ID NFT"
								FROM
									usuarios INNER JOIN usuario_tiene_nft
								ON
									usuario_tiene_nft.id_usuario = usuarios.id_usuario) as filtro_usuarios INNER JOIN nft
							ON 
								nft.id_nft = "ID NFT") as usuarios_con_nft INNER JOIN precio_nft
						ON
							"ID NFT" = precio_nft.id_nft
						ORDER BY
							"Fecha" DESC)

SELECT 
	"Usuario", SUM(precio) as "Dinero"
FROM
	fechas
WHERE
	"Fecha" = (SELECT MAX("Fecha") FROM fechas)
GROUP BY
	"Usuario"
FETCH FIRST 1 ROWS ONLY
```
